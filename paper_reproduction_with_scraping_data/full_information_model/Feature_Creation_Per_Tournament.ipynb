```python
import numpy as np
import pandas as pd
from timeit import default_timer as timer
from multiprocessing import Pool
import sys
import os

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path+"\\utilities")

from features_creation import features_creation
```

```python
data = pd.read_csv('../data/crawling_preprocess_data.csv',sep = ',')
```

```python
pd.options.display.max_columns = None
data.head(5)
```

```python
data['Match Date']=pd.to_datetime(data['Match Date'])
```

```python
data.shape
```

# Greek Basket League

```python
start = timer()
if __name__ ==  '__main__':
    pool = Pool(processes=6)  
    feature_greece = features_creation().run_features_creations(pool=pool,Data_Frame = data,
                                                                             Tournament = "Basket League",
                                                                             Date = "2015-8-01")
    pool.close()
    pool.terminate()
    pool.join()
end = timer()
print ("time:")
print ((end - start)/60)
```

```python
feature_greece.head()
```

```python
feature_greece.to_csv('Features_Per_Tournament/greece_features.csv',sep = ',',index = None, header=True)
```
