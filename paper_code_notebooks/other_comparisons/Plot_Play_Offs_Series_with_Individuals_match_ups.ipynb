```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
full_information_model_acc = [ 0.75, 0.75, 0.864, 0.667]
baseline_acc = [ 0.7, 0.75, 0.773, 0.714]
differences_acc_individuals = np.subtract(full_information_model_acc,baseline_acc)

baseline_acc = [ 0.75, 0.714, 0.750, 0.857]
full_information_model_acc = [ 0.75, 0.714, 1, 0.714]
differences_acc_series = np.subtract(full_information_model_acc,baseline_acc)

differences = [differences_acc_individuals,differences_acc_series]

title = '''Accuracy Differences of Best Predictive Model for
Full Information Models  with Baseline Vanilla Models
for Series and Individuals Play Offs Match-ups 2017-2018 '''



metrics = [ 'Play Offs Individuals Match-ups','Play Offs Series Match-ups' ]
tournaments = ["Euroleague","Eurocup","Greek Basket League","Liga ACB"]
colors = ['tab:blue','tab:orange']
#colors = ['tab:purple','tab:cyan','tab:olive']
#linestyles = ['-', '--', ':']
markers = ['v','^']

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)



for i in range(len(differences)):
    ax.plot(tournaments,differences[i],'o',label=metrics[i],color=colors[i], marker=markers[i],markersize=12)



ax.set_title(title, fontsize=30)
ax.set_ylabel('Differences Of Values', fontsize=20)
ax.set_xlabel('Tournaments', fontsize=20)
ax.legend(loc=1, prop={'size': 18})
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim(top=0.3)
ax.xaxis.set_tick_params(labelsize=20)
fig.tight_layout()
plt.savefig('baseline_vs_full_information_play_offs_series_individuals.png')
plt.show()
```
