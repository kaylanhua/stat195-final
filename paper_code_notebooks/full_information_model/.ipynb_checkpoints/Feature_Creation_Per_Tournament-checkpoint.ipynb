```python
import numpy as np
import pandas as pd
from timeit import default_timer as timer
from multiprocessing import Pool
import sys
import os

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path+"\\utilities")

from features_creation import features_creation
```

```python
data = pd.read_csv('../data/preprocess_data.csv',sep = ',')
```

```python
pd.options.display.max_columns = None
data.head(5)
```

```python
data['Match Date']=pd.to_datetime(data['Match Date'])
```

```python
data.shape
```

# Greek Basket League

```python
start = timer()
if __name__ ==  '__main__':
    pool = Pool(processes=6)  
    feature_greece = features_creation().run_features_creations(pool=pool,Data_Frame = data,
                                                                             Tournament = "Basket League",
                                                                             Date = "2014-8-01")
    pool.close()
    pool.terminate()
    pool.join()
end = timer()
print ("time:")
print ((end - start)/60)
```

```python
feature_greece.head()
```

```python
feature_greece.to_csv('Features_Per_Tournament/greece_features.csv',sep = ',',index = None, header=True)
```

# Sample Features

```python
data = data[data['Match Date']>'2014-10-01']

show_data_gr = data[data['Tournament'] == "Basket League"][['Tournament','Home Team','Away Team','Match Date','Phase']].reset_index(drop=True)
features_greece = pd.read_csv('Features_Per_Tournament/greece_features.csv').reset_index(drop=True)
show_data_gr = pd.concat([show_data_gr,features_greece],axis=1)

show_data_sp = data[data['Tournament'] == "Liga ACB"][['Tournament','Home Team','Away Team','Match Date','Phase']].reset_index(drop=True)
features_sp = pd.read_csv('Features_Per_Tournament/spain_features.csv').reset_index(drop=True)
show_data_sp = pd.concat([show_data_sp,features_sp],axis=1)

show_data_el = data[data['Tournament'] == "Euroleague"][['Tournament','Home Team','Away Team','Match Date','Phase']].reset_index(drop=True)
features_el = pd.read_csv('Features_Per_Tournament/euroleague_features.csv').reset_index(drop=True)
show_data_el = pd.concat([show_data_el,features_el],axis=1)

show_data_ec = data[data['Tournament'] == "Eurocup"][['Tournament','Home Team','Away Team','Match Date','Phase']].reset_index(drop=True)
features_ec = pd.read_csv('Features_Per_Tournament/eurocup_features.csv').reset_index(drop=True)
show_data_ec = pd.concat([show_data_ec,features_ec],axis=1)

show_data = pd.concat([show_data_gr,show_data_sp,show_data_el,show_data_ec]).reset_index(drop=True)
show_data['Match Date']=pd.to_datetime(show_data['Match Date'])

show_data = show_data[show_data['Match Date'].dt.month==10].groupby([show_data['Match Date'].dt.year,show_data['Tournament']]).head(1).sort_values(['Tournament','Match Date'])

#show_data.rename(columns=lambda c: 'Home ' + c if c.startswith('Team') else c)
```
